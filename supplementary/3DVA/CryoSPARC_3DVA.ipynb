### 10076 - use filter index, 256 pixels

```python
# This is form the directory of 3DVA 
data = np.load("J4_particles.cs")
```

```python
# Get the key for indexing
for i in range(len(data.dtype)):
    print(i, data.dtype.names[i], data[0][i])
```

```python
# We get the reaction coordernates here
test = np.array([x[43] for x in data])
for i in range(7):
    test = np.column_stack([test, np.array([x[45+2*i] for x in data])])
```

```python
test.shape
```

```python
test
```

```python
dens_emb = umap.UMAP(
    n_components=2,
    n_neighbors=15,
    min_dist=0.1,
    learning_rate=1e-3,
    densmap=True,
    dens_frac=0.3,
    dens_lambda=2.0,
    random_state=915
).fit_transform(test)
```

```python
plt.figure(figsize=(17,12))
plt.scatter(dens_emb[:,0], dens_emb[:,1], s=0.5, alpha=0.5);
```

```python
np.save("dens_3dva_10076.npy", dens_emb)
np.save("z_3dva_10076.npy", test)
```
